### 1.1.1 –ü—Ä–∏–º–µ—Ä

–ü—Ä–æ–≤–µ–¥–µ–º –ø—Ä—è–º—É—é ùë¶ = ùëöùë• + ùëè —á–µ—Ä–µ–∑ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–µ —Ç–æ—á–∫–∏.

```python
import numpy as np
x = np.array([0, 1, 2, 3])
y = np.array([0, 1, 2, 3])

# –ü–µ—Ä–µ–ø–∏—à–µ–º –ª–∏–Ω–µ–π–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ y = mx + c –∫–∞–∫ y = Ap, –≥–¥–µ A = [[ x 1]] –∏ p = [[m], [c]]
# –ü–æ—Å—Ç–æ—Ä–æ–∏–º A –ø–æ x

A = np.vstack([x, np.ones(len(x))]).T
print(A)

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ lstsq –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –µ–≥–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≤–µ–∫—Ç–æ—Ä–∞ p.
m, c = np.linalg.lstsq(A, y, rcond = None)[0]
print(m, c)

# –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –ø—Ä—è–æ–π –∏ —É–∫–∞–∂–µ–º –Ω–∞ –Ω–µ–º —Ç–æ—á–∫–∏.
import matplotlib.pyplot as plt
plt.plot(x, y, 'o', label='–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize=10)
plt.plot(x, m * x + c, 'r', label='–õ–∏–Ω–µ–π–Ω–∞—è —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è')
plt.legend()
plt.show()
```

### 1.1.2 –ü—Ä–∏–º–µ—Ä

–ü—É—Å—Ç—å ùë•, ùë¶ ‚Äì –≤–µ–∫—Ç–æ—Ä–∞ –¥–ª–∏–Ω–æ–π ùëõ > 3 (—Ç–æ—á–µ–∫ > 3). –ó–∞–¥–∞—á–∞ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤  
–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ —ç—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª–∏–Ω–æ–º–∞ –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ (–ø–∞—Ä–∞–±–æ–ª—ã). –¢–∞–∫–∏–º  
–æ–±—Ä–∞–∑–æ–º, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ —Ç–∞–∫–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ—Ä–∏–Ω–æ–º–∞ ùëé, ùëè, ùëê –ø–æ –º–µ—Ç–æ–¥—É  
–Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –î–∞–Ω–Ω—ã–µ –º—Ç–æ–≥—É—Ç –±—ã—Ç—å –ø–æ–ª—É—á–µ–Ω—ã –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ  
–∏–∑–º–µ—Ä–µ–Ω–∏–π. –ü–æ–∫–∞–∂–µ–º –ø—Ä–∏–º–µ—Ä –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –∏  
–∑–∞–≥—Ä—É–∑–∫–∏ –∏—Ö –∏–∑ —Ñ–∞–π–ª–∞.

```python
from numpy import *
from numpy.random import *
# –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ x –∏ y
delta = 1.0
x = linspace(-5, 5, 11)
y = x**2 + delta * (rand(11) - 0.5)
x += delta * (rand(11) - 0.5)

# –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ñ–∞–π–ª
x.tofile('x_data.txt', '\n')
y.tofile('y_data.txt', '\n')

# —á–∏—Ç–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ñ–∞–π–ª–æ–≤
x = fromfile('x_data.txt', float, sep='\n')
y = fromfile('y_data.txt', float, sep='\n')

print(x)
print(y)

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^2 + bx + c –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
# –∑–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**2, x, E]
m = vstack((x**2, x, ones(11))).T
# –Ω–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]

# –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5, 5]
x_prec = linspace(-5, 5, 101)
# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D')
# —Ä–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^2 + bx + c, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2]
plt.plot(x_prec, s[0] * x_prec**2 + s[1] * x_prec + s[2], '-', lw=2)
plt.grid()
plt.savefig('–ø–∞—Ä–∞–±–æ–ª–∞.png')
```

### 1.1.3 –ü—Ä–∏–º–µ—Ä

–ü–æ –¥–∞–Ω–Ω—ã–º –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—é –ø–æ–ª–∏–Ω–æ–º–∞  
—Ç—Ä–µ—Ç—å–µ–≥–æ –ø–æ—Ä—è–¥–∫–∞

```python
# –†–µ—à–µ–Ω–∏–µ
# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
# –∑–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**3, x, E]
m = vstack((x**3, x**2, x, ones(11))).T
# –Ω–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]

# –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5, 5]
x_prec = linspace(-5, 5, 101)
# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D')
# —Ä–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2]
plt.plot(x_prec, s[0] * x_prec**3 + s[1] * x_prec**2 + s[2]*x_prec + s[3], '-', lw=3)
plt.grid()
plt.savefig('–ø–æ–ª–∏–Ω–æ–º 3-–π —Å—Ç–µ–ø–µ–Ω–∏.png')
```

### –ó–∞–¥–∞–Ω–∏–µ

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —ç–∫—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—é –ø–æ–ª–∏–Ω–æ–º–∞–º–∏  
–ø–µ—Ä–≤–æ–π, –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Å—Ç–µ–ø–µ–Ω–∏.

```python
from numpy import *
from numpy.random import *
import matplotlib.pyplot as plt


x = array([0, 1, 2, 3, 4])
y = array([0, 1, 2, 3, 4])

# –ü–µ—Ä–µ–ø–∏—à–µ–º –ª–∏–Ω–µ–π–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ y = mx + c –∫–∞–∫ y = Ap, –≥–¥–µ A = [[ x 1]] –∏ p = [[m], [c]]
# –ü–æ—Å—Ç–æ—Ä–æ–∏–º A –ø–æ x

A = vstack([x, np.ones(len(x))]).T

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ lstsq –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –µ–≥–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≤–µ–∫—Ç–æ—Ä–∞ p.
m, c = linalg.lstsq(A, y, rcond = None)[0]

# –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –ø—Ä—è–æ–π –∏ —É–∫–∞–∂–µ–º –Ω–∞ –Ω–µ–º —Ç–æ—á–∫–∏.
plt.plot(x, y, 'o', label='–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize=10)
plt.plot(x, m * x + c, 'r', label='–õ–∏–Ω–µ–π–Ω–∞—è —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è')
plt.legend()
plt.grid()
plt.show()





# –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ x –∏ y
delta = 1.0
x = linspace(-10, 10, 15)
y = x**2 + delta * (rand(15) - 0.5)
x += delta * (rand(15) - 0.5)

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^2 + bx + c –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
# –∑–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**2, x, E]
m = vstack((x**2, x, ones(15))).T

# –Ω–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]
# –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [-10, 10]
x_prec = linspace(-10, 10, 1001)
# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D', label='–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize=10)
# —Ä–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^2 + bx + c, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2]
plt.plot(x_prec, s[0] * x_prec**2 + s[1] * x_prec + s[2], '-', lw=2,
         label='–≠–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è 2-–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.legend()
plt.grid()
plt.show()





# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
# –∑–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**3, x**2, x, E]
m = vstack((x**3, x**2, x, ones(15))).T
# –Ω–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]

# –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [-10, 10]
x_prec = linspace(-10, 10, 1001)
# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D', label='–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize=10)
# —Ä–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2]
plt.plot(x_prec, s[0] * x_prec**3 + s[1] * x_prec**2 + s[2]*x_prec + s[3], '-', lw=3,
         label='–≠–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è 3-–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.legend()
plt.grid()
plt.show()
```

### 1.1.4 –ü—Ä–∏–º–µ—Ä

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –Ω–∞—à–∏ —Ç–æ—á–µ—á–Ω–æ –∑–∞–¥–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ª–æ–∂–∏—Ç—Å—è  
–Ω–∞ –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ $ùëì(ùë•, ùëè) = ùëè_0 + ùëè_1ùëíùë•ùëù(‚àíùëè_2ùë•^2)$

```python
# –î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75, 0.5)
beta = (0.25, 0.75, 0.5)
def f(x, b0, b1, b2):
    return b0 + b1 * np.exp(-b2 * x**2)
# –∑–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)
# —Å–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
# –∑–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))

print(xdata)
print()
print(ydata)
print()

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ä–µ—à–µ–Ω–∏—è –≤ –≤–∏–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ f(x) –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö xdata –∏ ydata
from scipy.optimize import curve_fit
beta_opt, beta_cov = curve_fit(f, xdata, ydata)
print(beta_opt)

# –í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

# –í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

### 1.1.5 –ü—Ä–∏–º–µ—Ä

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –Ω–∞—à–∏ —Ç–æ—á–µ—á–Ω–æ –∑–∞–¥–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ª–æ–∂–∏—Ç—Å—è  
–Ω–∞ –∫—Ä–∏–≤—ã–µ –≤–∏–¥–∞:  
1) $ùëì(ùë•, ùëè) = ùëè_0 + ùëè_1ùë•$  
2) $ùëì(ùë•, ùëè) = ùëè_0 + ùëè_1ùë• + ùëè_2ùë•2$  
3) $ùëì(ùë•, ùëè) = ùëè_0 + ùëè_1ùëôùëõ(ùë•)$  
4) $ùëì(ùë•, ùëè) = ùëè_0 ùë•^{ùëè_1}$

–Ω–∞ ¬´–∞—Ç—Ä–∏–±—É—Ç—ã¬ª –∏ ¬´–º–µ—Ç–∫–∏¬ª. –ê—Ç—Ä–∏–±—É—Ç—ã ‚Äì —ç—Ç–æ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –∞ –º–µ—Ç–∫–∏  
‚Äì —ç—Ç–æ –∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã—Ö –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω—ã. –í  
–Ω–∞—à–µ–º –Ω–∞–±–æ—Ä–µ –≤—Å–µ–≥–æ –¥–≤–∞ —Å—Ç–æ–ª–±—Ü–∞ –∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å –æ—Ü–µ–Ω–∫—É –≤  
–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —á–∞—Å–æ–≤. –ß—Ç–æ–±—ã –∏–∑–≤–ª–µ—á—å –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏,  
–≤—ã–ø–æ–ª–Ω–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–π —Å–∫—Ä–∏–ø—Ç:

```python
x = dataset.iloc[:, : -1].values
y = dataset.iloc[:, 1].values
print(x)
print(y)

# –¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ —É –Ω–∞—Å –µ—Å—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∏—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã.  
# –ü—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π —Ñ—Ä–∞–≥–º–µ–Ω—Ç —Ä–∞–∑–¥–µ–ª—è–µ—Ç 80% –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á—é—â–∏–π –Ω–∞–±–æ—Ä, –∞ 20% –¥–∞–Ω–Ω—ã—Ö - –Ω–∞ –Ω–∞–±–æ—Ä —Ç–µ—Å—Ç–æ–≤
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# –¥–∞–ª–µ–µ –º–æ–∂–Ω–æ –æ–±—É—á–∏—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
# –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–ª–∞—Å—Å LinearRegression, —Å–æ–∑–¥–∞—Ç—å –µ–≥–æ —ç–∫–∑–µ–º–ø–ª—è—Ä –∏ –≤—ã–∑–≤–∞—Ç—å –º–µ—Ç–æ–¥ fit()
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# –ø—Ä–∏–≤–µ–¥–µ–º –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è –≤–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
print(regressor.intercept_)
print(regressor.coef_)
```

–ü–æ–ª—É—á–∏–≤—à–∏–π—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–æ–∂–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: —Å  
–∫–∞–∂–¥—ã–º –∑–∞—Ç—Ä–∞—á–µ–Ω–Ω—ã–º —á–∞—Å–æ–º –Ω–∞ –æ–±—É—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —ç–∫–∑–∞–º–µ–Ω–∞ –ø–æ–≤—ã—à–∞–µ—Ç—Å—è  
–ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ –Ω–∞ 17 –±–∞–ª–ª–æ–≤. –î–∞–ª–µ–µ –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã. –î–ª—è —ç—Ç–æ–≥–æ –º—ã  
–±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–∞—à–∏ —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ç–æ—á–Ω–æ –Ω–∞—à  
–∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—É—é –æ—Ü–µ–Ω–∫—É. –ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞  
—Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥:

```python
y_pred = regressor.predict(x_test)
# —Å—Ä–∞–≤–Ω–∏–º —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–º–∏
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –≤ –≤–∏–¥–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
df.plot(kind='bar')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

# –ø–æ—Å—Ç—Ä–æ–∏–º –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å —Ç–µ—Å—Ç–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
plt.scatter(x_test, y_test, color='gray')
plt.plot(x_test, y_pred, color='red', linewidth=2)
```

### –ó–∞–¥–∞–Ω–∏–µ

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–≤—É—Ö  
—Å—Ç–æ–ª–±—Ü–æ–≤. –î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –º–æ–∂–Ω–æ –≤–∑—è—Ç—å —Ç–æ—á–µ—á–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç—ã –æ—Ç  
–æ–ø—ã—Ç–∞ —Ä–∞–±–æ—Ç—ã:  
(https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv).  
–ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑.

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
# –∏—Å–ø–æ–ª—å–∑—É–µ–º pandas –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö,
# matplotlib –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∏ sklearn –¥–ª—è –æ–±—É—á–µ–Ω–∏—è
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# –∑–∞–≥—Ä—É–∑–∏–º –¥–∞–Ω–Ω—ã–µ
url = 'https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-\
regression/master/Salary_Data.csv'
dataset = pd.read_csv(url)

# –ù–∞—Ä–∏—Å—É–µ–º —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
plt.scatter(dataset['YearsExperience'], dataset['Salary'], color='b')
plt.xlabel("–ß–∏—Å–ª–æ –ª–µ—Ç")
plt.ylabel("–ó–∞—Ä–ø–ª–∞—Ç–∞")
plt.show()

# —Ä–∞–∑–¥–µ–ª–∏–º –∏–Ω—Ñ–æ—Ä–º—Ü–∏—é –Ω–∞ –∞—Ç—Ä–∏–±—É—Ç—ã(–æ–ø—ã—Ç —Ä–∞–±–æ—Ç—ã) –∏ –º–µ—Ç–∫–µ(–∑–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞)
x = dataset.iloc[:, : -1].values
y = dataset.iloc[:, 1].values

# –¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ —É –Ω–∞—Å –µ—Å—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∏—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã.  
# –ü—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π —Ñ—Ä–∞–≥–º–µ–Ω—Ç —Ä–∞–∑–¥–µ–ª—è–µ—Ç 80% –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á—é—â–∏–π –Ω–∞–±–æ—Ä, –∞ 20% –¥–∞–Ω–Ω—ã—Ö - –Ω–∞ –Ω–∞–±–æ—Ä —Ç–µ—Å—Ç–æ–≤
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# –¥–∞–ª–µ–µ –º–æ–∂–Ω–æ –æ–±—É—á–∏—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
# –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–ª–∞—Å—Å LinearRegression, —Å–æ–∑–¥–∞—Ç—å –µ–≥–æ —ç–∫–∑–µ–º–ø–ª—è—Ä –∏ –≤—ã–∑–≤–∞—Ç—å –º–µ—Ç–æ–¥ fit()
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å –ø–æ–º–æ—â—å—é –Ω–∞—à–µ–π –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
y_pred = regressor.predict(x_test)
# —Å—Ä–∞–≤–Ω–∏–º —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–º–∏
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –≤ –≤–∏–¥–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
df.plot(kind='bar')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

# –ø–æ—Å—Ç—Ä–æ–∏–º –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å —Ç–µ—Å—Ç–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
plt.scatter(x_test, y_test, color='gray')
plt.plot(x_test, y_pred, color='red', linewidth=2)
```

### 1.3.1 –ü—Ä–∏–º–µ—Ä

–î–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –º–æ–∂–Ω–æ –∑–∞–¥–µ–π—Å—Ç–≤–æ–≤–∞—Ç—å —É–∂–µ  
–∏–∑–≤–µ—Å—Ç–Ω—ã–π –º–µ—Ç–æ–¥ numpy.linalg.lstsq.

```python
import numpy as np

y = [1, 2, 3, 4, 3, 4, 5, 3, 5, 5, 4, 5, 4, 5, 4, 5, 6, 0, 6, 3, 1, 3, 1]
x = [[0, 2, 4, 1, 5, 4, 5, 9, 9, 9, 3, 7, 8, 8, 6, 6, 5, 5, 5, 6, 6, 5, 5],
     [4, 1, 2, 3, 4, 5, 6, 7, 5, 8, 7, 8, 7, 8, 7, 8, 6, 8, 9, 2, 1, 5, 6],
     [4, 1, 2, 5, 6, 7, 8, 9, 7, 8, 7, 8, 7, 4, 3, 1, 2, 3, 4, 1, 3, 9, 7]]
x = np.transpose(x) # transpose these input vectors
x = np.c_[x, np.ones(x.shape[0])] # add bias term
linreg = np.linalg.lstsq(x, y, rcond=None)[0]
print(linreg)
```

### 1.3.2 –ü—Ä–∏–º–µ—Ä

–î–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–∞–¥–∞—á–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –º–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π  
–ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å—Ä–µ–¥—Å—Ç–≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ s—Åikit-learn.

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics


y = [1, 2, 3, 4, 3, 4, 5, 3, 5, 5, 4, 5, 4, 5, 4, 5, 6, 0, 6, 3, 1, 3, 1]
x = [[0, 2, 4, 1, 5, 4, 5, 9, 9, 9, 3, 7, 8, 8, 6, 6, 5, 5, 5, 6, 6, 5, 5],
     [4, 1, 2, 3, 4, 5, 6, 7, 5, 8, 7, 8, 7, 8, 7, 8, 6, 8, 9, 2, 1, 5, 6],
     [4, 1, 2, 5, 6, 7, 8, 9, 7, 8, 7, 8, 7, 4, 3, 1, 2, 3, 4, 1, 3, 9, 7]]

# —Ñ–æ—Ä–º–∏—Ä—É–µ–º DataFrame –∏–∑ –¥–≤—É—Ö —Å–ø–∏—Å–∫–æ–≤
new_y = np.array(y)
new_y = new_y.transpose()
df1 = pd.DataFrame(new_y)
new_x = np.array(x)
new_x = new_x.transpose()
df2 = pd.DataFrame(new_x)
df1 = df1.rename(columns={0: 'y'}, inplace=False)
df2 = df2.rename(columns={0: 'x1', 1: 'x2', 2: 'x3'}, inplace=False)

frames = [df1, df2]
dataset = pd.concat([df1, df2], axis=1, join="inner")
print(dataset.head())
print()

# –∏–∑—É—á–∏–º –¥–∞–Ω–Ω—ã–µ
print(dataset.shape)
print()
print(dataset.describe())
print()

# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –º–µ—Ç–∫–∏ –∏ –∞—Ç—Ä–∏–±—É—Ç—ã
x = dataset[['x1', 'x2', 'x3']]
y = dataset['y']

# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤—ã–ø–æ–ª–Ω—è–µ–º —Ç–æ—Ç –∂–µ –∫–æ–¥, —á—Ç–æ –∏ —Ä–∞–Ω—å—à–µ, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ fit() –∫–ª–∞—Å—Å–∞ LinearRegression
regressor = LinearRegression()
print(regressor.fit(x_train, y_train))
print()

# –≤—ã–≤–µ–¥–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏
coeff_df = pd.DataFrame(regressor.coef_, x.columns, columns=['Coefficient'])
print(coeff_df)
print()

# –ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥
y_pred = regressor.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)
print()

# –ü–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–≥ - –æ—Ü–µ–Ω–∏—Ç—å –ø—Ä–æ–∏–∑–≤–æ–≤–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ú—ã —Å–¥–µ–ª–∞–µ–º —ç—Ç–æ, –Ω–∞–π–¥—è –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è MSE
print('Mean Square Error:', metrics.mean_squared_error(y_test, y_pred))
```

### –ó–∞–¥–∞–Ω–∏–µ

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö  
–¥–∞–Ω–Ω—ã—Ö –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤. –î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –º–æ–∂–Ω–æ –≤–∑—è—Ç—å –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è  
–≥–∞–∑–∞ (–≤ –º–∏–ª–ª–∏–æ–Ω–∞—Ö –≥–∞–ª–ª–æ–Ω–æ–≤) –≤ 48 —à—Ç–∞—Ç–∞—Ö –°–®–ê –∏–ª–∏ –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –æ  
–∫–∞—á–µ—Å—Ç–≤–µ –∫—Ä–∞—Å–Ω–æ–≥–æ –≤–∏–Ω–∞ (1) –∏ (2) —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã  
–º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑.
1. https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv
2. https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# –∑–∞–≥—Ä—É–∑–∏–º –¥–∞–Ω–Ω—ã–µ
url1 = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/\
data/petrol_consumption.csv'
url2 = 'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/\
master/winequality-red.csv'
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)

# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –º–µ—Ç–∫–∏ –∏ –∞—Ç—Ä–∏–±—É—Ç—ã
x1 = df1[['Petrol_tax', 'Average_income', 'Paved_Highways', 'Population_Driver_licence(%)']]
y1 = df1['Petrol_Consumption']
x2 = df2[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
          'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y2 = df2['quality']

# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2, random_state=0)
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.2, random_state=0)

# –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤—ã–ø–æ–ª–Ω—è–µ–º —Ç–æ—Ç –∂–µ –∫–æ–¥, —á—Ç–æ –∏ —Ä–∞–Ω—å—à–µ, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ fit() –∫–ª–∞—Å—Å–∞ LinearRegression
regressor1 = LinearRegression()
regressor1.fit(x1_train, y1_train)
regressor2 = LinearRegression()
regressor2.fit(x2_train, y2_train)

# –≤—ã–≤–µ–¥–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏
coeff_df1 = pd.DataFrame(regressor1.coef_, x1.columns, columns=['Coefficient'])
print(coeff_df1)
print()
coeff_df2 = pd.DataFrame(regressor2.coef_, x2.columns, columns=['Coefficient'])
print(coeff_df2)
print()

# –ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥
y1_pred = regressor1.predict(x1_test)
df1 = pd.DataFrame({'Actual': y1_test, 'Predicted': y1_pred})
print(df1)
print()
y2_pred = regressor2.predict(x2_test)
df2 = pd.DataFrame({'Actual': y2_test, 'Predicted': y2_pred})
print(df2)
print()

# –ü–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–≥ - –æ—Ü–µ–Ω–∏—Ç—å –ø—Ä–æ–∏–∑–≤–æ–≤–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ú—ã —Å–¥–µ–ª–∞–µ–º —ç—Ç–æ, –Ω–∞–π–¥—è –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è MSE
print('Mean Square Error:', metrics.mean_squared_error(y1_test, y1_pred))
print()
print('Mean Square Error:', metrics.mean_squared_error(y2_test, y2_pred))
```

–≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ –ø–æ–ª—É—á–µ–Ω—ã N ‚àí –∑–Ω–∞—á–µ–Ω–∏–π –≤–µ–ª–∏—á–∏–Ω—ã Y –ø—Ä–∏  
—Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –≤–µ–ª–∏—á–∏–Ω—ã X. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª–∏–Ω–æ–º—ã –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π  
—Å—Ç–µ–ø–µ–Ω–∏, –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞, —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º  
–º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã–≤–æ–¥—è—Ç—Å—è –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü  
–∑–Ω–∞—á–µ–Ω–∏–π –∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª–∏–Ω–æ–º–æ–≤.

|3,0|2,0|

```python
import numpy as np
import matplotlib.pyplot as plt

# –∑–∞–¥–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
x = np.array([2.0, 2.2, 2.4, 2.6, 2.8, 3.0])
y = np.array([4.0, 2.0, 4.0, 2.0, 5.0, 2.0])

# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ—Ä–µ–Ω–∏–π
n = len(x)

# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ø–æ–ª–∏–Ω–æ–º–∞ –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏
a1, a0 = np.polyfit(x, y, 1)

# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ø–æ–ª–∏–Ω–æ–º–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏
A = np.vstack([x**2, x, np.ones(n)]).T
b2, b1, b0 = np.linalg.lstsq(A, y, rcond=None)[0]

# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π Y
y1 = a0 + a1*x
y2 = b0 + b1*x + b2*x**2

# –≤—ã–≤–æ–¥ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ø–æ–ª–∏–Ω–æ–º–æ–≤
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ–ª–∏–Ω–æ–º–∞ –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏: a1 = {a1:.3f}, a0 = {a0:.3f}")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ–ª–∏–Ω–æ–º–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏: b2 = {b2:.3f}, b1 = {b1:.3f}, b0 = {b0:.3f}")

# –≤—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã –∑–Ω–∞—á–µ–Ω–∏–π X –∏ Y
print("X\tY\tY1\tY2")
for i in range(n):
    print(f"{x[i]:.1f}\t{y[i]:.1f}\t{y1[i]:.3f}\t{y2[i]:.3f}")

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.scatter(x, y, label='–ò–∑–º–µ—Ä–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
plt.plot(x, y1, 'r', label='–ü–æ–ª–∏–Ω–æ–º –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.plot(x, y2, 'g', label='–ü–æ–ª–∏–Ω–æ–º –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.legend(loc='best')
plt.show()
```
