### 4.2.2 Пример

Для функции $sin(xe^{\cos(x)})$ на интервале [0;10] с шагом 0,05 построить  
график функции, закрасить площадь под ним и вычислить значение  
этой площади.  
Для вычисления площади воспользуйте функции trapz(), вычисляющей  
интеграл по правилу трапеции. Для ее корректной работы необходимо  
подключить следующие библиотеки:  
from scipy.integrate import simps  
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz
x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x * np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c = "r")
plt.fill_between(x, y)

area = trapz(y)
print(area)
```

```python
from random import random as rand
from statistics import median as med, mean

x = [rand() for i in range(10)]
av = mean(x)
me = med(x)

print('Среднее: {}'.format(av))
print('Медиана: {}'.format(me))

if av > me:
    print('Среднее > Медиана')
elif av < me:
    print('Среднее < Медиана')
else:
    print('Среднее == Медиана')

plt.grid()
plt.scatter(range(10), x)
```

### 4.3.2 Задание

Дана функция $\frac{\sqrt{1+e^\sqrt x+\cos x^2}}{|1-sin^3 x|}+\ln |2x|$ Создать массив из 10 значений  
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой  
половины массива и построить графики для основного массива –  
линейный и для среза – точечный

```python
from math import sqrt, exp, cos, sin

y = [sqrt(1 + exp(sqrt(x)) + cos(x*x)) / abs(1 - pow(sin(x), 3))
     for x in range(10)]

plt.grid()
plt.plot(range(10), y)

plt.show()
plt.grid()
plt.scatter(range(5), y[: 5])
```

### 4.3.4 Задание

Откройте статистические данные по стоимости акций Apple,  
Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на  
плоскости и оцените их динамику. Подумайте, как можно улучшить  
читаемость результатов.  
Для начальных данных допустимо использовать следующий ресурс:  
https://finance.yahoo.com

```python
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ["AAPL", "MSFT", "GOOGL"]
start_date = "2021-01-01"
end_date = "2021-12-31"

data = yf.download(tickers, start=start_date, end=end_date)["Adj Close"]

fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=True)

for i, ticker in enumerate(tickers):
    axs[i].plot(data.index, data[ticker], label=ticker)
    axs[i].set_title(ticker)
    axs[i].set_ylabel("Adjusted Close Price ($)")
    axs[i].legend(loc="upper left")

plt.xlabel("Date")
plt.suptitle("Stock Prices for Apple, Microsoft, and Google in 2021")
plt.show()
```

### 4.3.5 Задание

Создайте простейший калькулятор, включающий основные действия  
для двух переменных ′+′, ′−′, ′∙′, ′÷′, а также вычисление  
следующих функций: $e^{x+y}, \sin{(x+y)}, \cos{(x+y)}, x^y.$

```python
from math import exp, sin, cos

ex = input()
ex.replace(' ', '')

if ex.find('exp') > -1:
    print(exp(float(ex[4: ex.find('+')]) + float(ex[ex.find('+') + 1: -1])))
elif ex.find('sin') > -1:
    print(sin(float(ex[4: ex.find('+')]) + float(ex[ex.find('+') + 1: -1])))
elif ex.find('cos') > -1:
    print(cos(float(ex[4: ex.find('+')]) + float(ex[ex.find('+') + 1: -1])))
elif ex.find('pow') > -1:
    print(pow(float(ex[4: ex.find(',')]), float(ex[ex.find(',') + 1: -1])))
elif ex.find('+') > -1:
    print(float(ex[: ex.find('+')]) + float(ex[ex.find('+') + 1: ]))
elif ex.find('-') > -1:
    print(float(ex[: ex.find('-')]) - float(ex[ex.find('-') + 1: ]))
elif ex.find('*') > -1:
    print(float(ex[: ex.find('*')]) * float(ex[ex.find('*') + 1: ]))
elif ex.find('/') > -1:
    print(float(ex[: ex.find('/')]) / float(ex[ex.find('/') + 1: ]))
```
